```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

# Load the datasets

data_2022 = pd.read_csv('Cleaned_2022_2000-miler_roster_Python_Direction.csv', encoding='ISO-8859-1')
data_2023 = pd.read_csv('Cleaned_2023_2000-miler_roster_Python_Direction.csv', encoding='ISO-8859-1')
data_2024 = pd.read_csv('Cleaned_2024_2000-miler_roster_Python_Direction.csv', encoding='ISO-8859-1')

# Add a year column to each dataframe
data_2022['Year'] = 2022
data_2023['Year'] = 2023
data_2024['Year'] = 2024

# Combine the datasets into a single dataframe
combined_data = pd.concat([data_2022, data_2023, data_2024], ignore_index=True)

# Frequency analysis
frequency_data = combined_data.groupby(['Year', 'Direction']).size().unstack().fillna(0)

# Select only the directions of interest
directions_of_interest = ['Northbounders', 'Southbounders']
frequency_data = frequency_data[directions_of_interest]

# Plotting the bar chart with specified colors
ax = frequency_data.plot(kind='bar', figsize=(12, 8), color=['darkblue', 'brown', 'green'])
plt.title('2,000-Miler Direction per Year')
plt.xlabel('Year')
plt.ylabel('2,000-Milers')
plt.legend(title='Direction')
plt.xticks(rotation=0)

# Show the plot
plt.show()

# Chi-square test
chi2, p, dof, expected = chi2_contingency(frequency_data)

print(f"Chi-square Statistic: {chi2}")
print(f"P-value: {p}")

# If p-value is less than 0.05, we reject the null hypothesis
if p < 0.05:
    print("There is a significant association between year and direction.")
else:
    print("There is no significant association between year and direction.")
```
