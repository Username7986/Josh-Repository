#### 2000-miler by decade

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Load the cleaned dataset
file_path = 'cleaned_2000-milers_by_decade.csv'
data = pd.read_csv(file_path)

# Create a line graph
plt.figure(figsize=(10, 6))
plt.plot(data['Decade'], data['Milestone'], marker='o', linestyle='-', color='b')
plt.title('2,000 Milers by Decade')
plt.xlabel('Decade')
plt.ylabel('Number of 2,000 Milers')
plt.grid(True)
plt.show()

# Perform trend analysis
data['Const'] = sm.add_constant(data['Decade'])
model = sm.OLS(data['Milestone'], data[['Const', 'Decade']]).fit()
trend = model.predict(data[['Const', 'Decade']])

# Plot the trend line
plt.figure(figsize=(10, 6))
plt.plot(data['Decade'], data['Milestone'], marker='o', linestyle='-', color='b', label='Actual')
plt.plot(data['Decade'], trend, linestyle='--', color='r', label='Trend')
plt.title('Trend Analysis of 2,000 Milers by Decade')
plt.xlabel('Decade')
plt.ylabel('Number of 2,000 Milers')
plt.legend()
plt.grid(True)
plt.show()

# Perform time series analysis
data.set_index('Decade', inplace=True)
decomposition = sm.tsa.seasonal_decompose(data['Milestone'], model='additive', period=1)

# Plot the decomposed time series
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 12))
decomposition.trend.plot(ax=ax1)
ax1.set_title('Trend')
ax1.set_ylabel('Number of 2,000 Milers')

decomposition.seasonal.plot(ax=ax2)
ax2.set_title('Seasonal')
ax2.set_ylabel('Number of 2,000 Milers')

decomposition.resid.plot(ax=ax3)
ax3.set_title('Residual')
ax3.set_ylabel('Number of 2,000 Milers')

plt.tight_layout()
plt.show()


```

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Load the cleaned dataset
file_path = 'cleaned_2000-milers_by_decade.csv'
data = pd.read_csv(file_path)

# Create a line graph
plt.figure(figsize=(10, 6))
plt.plot(data['Decade'], data['Milestone'], marker='o', linestyle='-', color='b')
plt.title('2,000 Milers by Decade')
plt.xlabel('Decade')
plt.ylabel('Number of 2,000 Milers')
plt.grid(True)
plt.show()

# Perform trend analysis
X = sm.add_constant(data['Decade'])
model = sm.OLS(data['Milestone'], X).fit()
trend = model.predict(X)

# Project the number of 2,000 Milers for the 2020s
projection_decade = pd.DataFrame({'Decade': [2020]})
projection_decade = sm.add_constant(projection_decade)
projection = model.predict(projection_decade)

# Combine actual data and projection for plotting
projection_data = pd.DataFrame({'Decade': [2020], 'Milestone': projection})

# Plot the trend line and projection
plt.figure(figsize=(10, 6))
plt.plot(data['Decade'], data['Milestone'], marker='o', linestyle='-', color='b', label='Actual')
plt.plot(data['Decade'], trend, linestyle='--', color='r', label='Trend')
plt.plot(projection_data['Decade'], projection_data['Milestone'], marker='o', linestyle='-', color='g', label='Projection (2020s)')
plt.title('Trend Analysis and Projection of 2,000 Milers by Decade')
plt.xlabel('Decade')
plt.ylabel('Number of 2,000 Milers')
plt.legend()
plt.grid(True)
plt.show()

# Perform time series analysis
data.set_index('Decade', inplace=True)
decomposition = sm.tsa.seasonal_decompose(data['Milestone'], model='additive', period=1)

# Plot the decomposed time series
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 12))
decomposition.trend.plot(ax=ax1)
ax1.set_title('Trend')
ax1.set_ylabel('Number of 2,000 Milers')

decomposition.seasonal.plot(ax=ax2)
ax2.set_title('Seasonal')
ax2.set_ylabel('Number of 2,000 Milers')

decomposition.resid.plot(ax=ax3)
ax3.set_title('Residual')
ax3.set_ylabel('Number of 2,000 Milers')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Load the newly uploaded dataset
file_path = 'redo_2000-miler_by_decade.csv'
data = pd.read_csv(file_path)

# Transform the data to long format
data_long = data.melt(var_name='Decade', value_name='Milestone')
data_long['Decade'] = data_long['Decade'].str.extract('(\d+)').astype(int)

# Create a line graph
plt.figure(figsize=(10, 6))
plt.plot(data_long['Decade'], data_long['Milestone'], marker='o', linestyle='-', color='b')
plt.title('2,000 Milers by Decade')
plt.xlabel('Decade')
plt.ylabel('Number of 2,000 Milers')
plt.grid(True)
plt.show()

# Perform trend analysis
X = sm.add_constant(data_long['Decade'])
model = sm.OLS(data_long['Milestone'], X).fit()
trend = model.predict(X)

# Project the number of 2,000 Milers for the 2020s
projection_decade = pd.DataFrame({'const': [1], 'Decade': [2020]})
projection = model.predict(projection_decade)
print("Projected number of 2,000 Milers for 2020s:", projection[0])

# Combine actual data and projection for plotting
projection_data = pd.DataFrame({'Decade': [2020], 'Milestone': projection})

# Plot the trend line and projection
plt.figure(figsize=(10, 6))
plt.plot(data_long['Decade'], data_long['Milestone'], marker='o', linestyle='-', color='b', label='Actual')
plt.plot(data_long['Decade'], trend, linestyle='--', color='r', label='Trend')
plt.scatter(projection_data['Decade'], projection_data['Milestone'], color='g', s=100, label='Projection (2020s)')
plt.title('Trend Analysis and Projection of 2,000 Milers by Decade')
plt.xlabel('Decade')
plt.ylabel('Number of 2,000 Milers')
plt.legend()
plt.grid(True)
plt.show()

# Perform time series analysis
data_long.set_index('Decade', inplace=True)
decomposition = sm.tsa.seasonal_decompose(data_long['Milestone'], model='additive', period=1)

# Plot the decomposed time series
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(10, 12))
decomposition.trend.plot(ax=ax1)
ax1.set_title('Trend')
ax1.set_ylabel('Number of 2,000 Milers')

decomposition.seasonal.plot(ax=ax2)
ax2.set_title('Seasonal')
ax2.set_ylabel('Number of 2,000 Milers')

decomposition.resid.plot(ax=ax3)
ax3.set_title('Residual')
ax3.set_ylabel('Number of 2,000 Milers')

plt.tight_layout()
plt.show()

```
